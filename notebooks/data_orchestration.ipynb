```python
from google.cloud import bigquery

# Initialize BigQuery client
client = bigquery.Client(project="big-data-assignment4-463612")

# Load and run a SQL file
def run_sql_script(file_path):
    with open(file_path, "r") as file:
        sql_query = file.read()
    job = client.query(sql_query)
    job.result()  # Wait for job to finish
    print(f"✅ Executed: {file_path}")

# Run all scripts
base_path = '/content/drive/My Drive/bigdata/project/sql'
sql_scripts = [
    base_path+'/dim_team.sql',
    base_path+'/dim_player.sql',
    base_path+'/dim_match.sql',
    base_path+'/fact_deliveries.sql'
]

for script in sql_scripts:
    run_sql_script(script)
```
